## **No. Of Rapes Per Year**

```python
df.groupby('Year')['Rape'].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(30,8))
sns.barplot(x='State',y='Rape',data=df)
plt.xticks(rotation=75)
plt.title("No. Of Rapes")
plt.show()
```
