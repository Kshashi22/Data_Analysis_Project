## **Average Annual Rainfall**

```python
df.groupby('Crop_Year')['Annual_Rainfall'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Crop_Year',y='Annual_Rainfall',data=df)
plt.xticks(rotation=75)
plt.title("Annual_Rainfall")
plt.show()
```

## **Number of Production per Year**

```python
df['Crop_Year'].value_counts()
```

```python
sns.countplot(x='Crop_Year',data=df)
plt.xticks(rotation=75)
plt.title("Number of Production Per Year")
plt.show()
```
